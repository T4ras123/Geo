```python
import cartopy.crs as ccrs
import geopandas as gpd

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

fig, ax = plt.subplots(subplot_kw={'projection': ccrs.Robinson()})
world.plot(ax=ax, transform=ccrs.PlateCarree())
plt.show()

fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()})
world.plot(ax=ax, transform=ccrs.PlateCarree())
plt.show()

fig, ax = plt.subplots(subplot_kw={'projection': ccrs.AzimuthalEquidistant()})
world.plot(ax=ax, transform=ccrs.PlateCarree())
plt.show()

fig, ax = plt.subplots(subplot_kw={'projection': ccrs.Mercator()})
world.plot(ax=ax, transform=ccrs.PlateCarree())
plt.show()

fig, ax = plt.subplots(subplot_kw={'projection': ccrs.Miller()})
world.plot(ax=ax, transform=ccrs.PlateCarree())
plt.show()

fig, ax = plt.subplots(subplot_kw={'projection': ccrs.Mollweide()})
world.plot(ax=ax, transform=ccrs.PlateCarree())
plt.show()

fig, ax = plt.subplots(subplot_kw={'projection': ccrs.Sinusoidal()})
world.plot(ax=ax, transform=ccrs.PlateCarree())
plt.show()

fig, ax = plt.subplots(subplot_kw={'projection': ccrs.InterruptedGoodeHomolosine()})
world.plot(ax=ax, transform=ccrs.PlateCarree())
plt.show()
```
