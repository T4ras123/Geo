```python
import matplotlib.pyplot as plt

logo = plt.imread('resources/globe_black.png')
```

```python
import geopandas as gpd

data = gpd.read_file("../PythonMapsData/NaturalEarth/10m_cultural/ne_10m_time_zones.shp")
countries = gpd.read_file("../PythonMapsData/NaturalEarth/10m_cultural/ne_10m_admin_0_countries.shp")
```

```python
print(data.columns)
print(data.zone.value_counts())
```

```python
data = data.sort_values('name')
data['name'] = data['name'].astype(float)
```

```python
main_windows = data.map_color6.unique().tolist()
```

```python
x = data.name.unique().tolist()
x.sort()
print(x)
```

```python
from matplotlib import cm
import numpy as np
import pandas as pd
import matplotlib

purples = cm.get_cmap('tab20c', 40)
newcolors = purples(np.linspace(0, 1, 40))
colors = []
for i in range(40):
    rgb = newcolors[i][:3]  # will return rgba, we take only first 3 so we get rgb
    colors.append(matplotlib.colors.rgb2hex(rgb))
colors = pd.DataFrame({'name': x, 'color': colors})
data = pd.merge(data, colors, on='name')
```

```python
new_data = gpd.sjoin(data, countries, op='intersects')
```

```python
print(new_data)
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

fig, ax = plt.subplots(facecolor='#FCF6F5FF', subplot_kw={'projection': ccrs.PlateCarree()})
fig.set_size_inches(7, 3.5)
ax.set_facecolor('#FCF6F5FF')
new_data.plot(ax=ax, color='#FCF6F5FF', edgecolor='black', lw=1, transform=ccrs.PlateCarree())
#data.plot(ax=ax, color=data['color'], edgecolor='black', transform=ccrs.PlateCarree(), alpha=0.8)

plt.setp(ax.spines.values(), color='#FCF6F5FF')
plt.setp([ax.get_xticklines(), ax.get_yticklines()], color='#FCF6F5FF')
newax = fig.add_axes([0.82, 0.7, 0.08, 0.08], anchor='NE')
newax.imshow(logo)
newax.axis('off')
ax.set_ylim(-60, 90)
plt.savefig(f"boundaries.png", dpi=200, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(facecolor='#FCF6F5FF', subplot_kw={'projection': ccrs.PlateCarree()})
fig.set_size_inches(7, 3.5)
countries.plot(ax=ax, color='none', edgecolor='black', lw=1)
data.plot(ax=ax, color=data['color'], edgecolor='white', lw=1, transform=ccrs.PlateCarree(), alpha=0.6)

plt.setp(ax.spines.values(), color='#FCF6F5FF')
plt.setp([ax.get_xticklines(), ax.get_yticklines()], color='#FCF6F5FF')
newax = fig.add_axes([0.82, 0.04, 0.08, 0.08], anchor='NE')
newax.imshow(logo)
newax.axis('off')
txt = ax.text(0.0, -0.08, "Time Zones \n@PythonMaps",
              size=4,
              color='grey',
              transform = ax.transAxes)
plt.savefig(f"Polygon.png", dpi=600, bbox_inches='tight')
plt.show()
```
